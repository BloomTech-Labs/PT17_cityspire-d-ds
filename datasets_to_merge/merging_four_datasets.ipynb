```python
import numpy as np
import pandas as pd

df_census = pd.read_csv('census_by_city.csv')    # (25235, 35)
df_crime = pd.read_csv('fbi_crime_uscities.csv') # (8103, 16)
df_rent = pd.read_csv('rent_cleaned.csv')        # (2831, 4)
df_poll = pd.read_csv('pollution_summary.csv')   # (845, 20)

# replace 'State_Id' in df_census so it matches the other columns
df_census.rename(columns = {'State_Id':'State'}, inplace = True)

# put all the dfs in a list
df_lst = [df_census, df_crime, df_rent, df_poll]

# remove whitespace in all the dfs
for x in df_lst:
    x['City'] = x['City'].str.strip()
    x['State'] = x['State'].str.strip()

# make sure it worked    
for x in df_lst:
    print(x['City'][0])
    print(x['State'][0])
```

```python
# merging the data and following its shape sequentially
df_net = pd.merge(df_census, df_crime, on=['City', 'State'])
print(df_net.shape)
df_net2 = pd.merge(df_net, df_rent, on=['City', 'State'])
print(df_net2.shape)
df_net3 = pd.merge(df_net2, df_poll, on=['City', 'State'])
print(df_net3.shape)
df_net3
```
