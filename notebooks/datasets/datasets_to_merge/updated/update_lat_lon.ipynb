```python
import pandas as pd
import numpy as np
import os

pd.set_option('display.max_columns', None)
```

```python
nn_model = pd.read_csv('https://raw.githubusercontent.com/Lambda-School-Labs/labspt15-cityspire-g-ds/main/notebooks/model/nearest_neighbor/nn_model.csv')
```

```python
print(nn_model.shape)
nn_model.head()
```

```python
lat_lon = pd.read_csv('https://raw.githubusercontent.com/kelvins/US-Cities-Database/main/csv/us_cities.csv')
```

```python
print(lat_lon.shape)
lat_lon.head()
```

```python
lat_lon.rename(columns = {'STATE_CODE':'State', 'CITY': 'City'}, inplace = True)
lat_lon = lat_lon[['City', 'State', 'LATITUDE', 'LONGITUDE']]
```

```python
nn_model = pd.merge(nn_model, lat_lon, on=['City', 'State'])
```

```python
print(nn_model.shape)
nn_model.head()
```

```python
nn_model[nn_model.duplicated(subset=['City', 'State', 'lat', 'lon'], keep=False)]
```

```python
nn_model = nn_model.drop_duplicates(subset=['City', 'State', 'lat', 'lon'], keep='first')
```

```python
nn_model.drop(['lat', 'lon'], axis=1, inplace=True)
nn_model.rename(columns = {'LATITUDE':'lat', 'LONGITUDE': 'lon'}, inplace = True)
print(nn_model.shape)
nn_model.head()
```

```python
nn_model.to_csv('final.csv', index = False) 
```
