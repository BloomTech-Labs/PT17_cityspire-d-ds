```python
import pandas as pd
```

```python
#reading cities list from previous code base downloaded as csv
city = pd.read_csv('final.csv')
```

```python
city_names = city[['City','State','lat','lon']]
```

```python
city_names.head()
```

```python
#using weather station list to find out closest stations to cities list.
stations = pd.read_csv('stations.csv')
```

```python
stations.head()
```

```python
#merging station list with cities to add weather station numbers to relevant cities.
cities_with_stations = pd.merge(city_names, stations,  how='left', left_on=['City','State'], right_on = ['City','State'])
```

```python
cities_with_stations.head(200)
```

```python
cities_with_stations.shape
```

```python
#dropping duplicate station numbers as some cities can have multiple stations
#this is done for simplicity.
cities_final1 = cities_with_stations.drop_duplicates(subset='StationNumber')
```

```python
cities_final1.shape
```

```python
cities_final1.head(200)
```

```python
cities_final1['StationNumber'].value_counts()
```

```python
#cleaning data.
cities_final1['StationNumber'].dropna()
```

```python
cities_final1.shape
```

```python
cities_final1.info()
```

```python
cities_final1.head()
```

```python
#999999 indicates no station number for that city and hence being dropped.
cities_final2 = cities_final1[cities_final1["StationNumber"] != '999999']
```

```python
cities_final2  = cities_final2.dropna()
```

```python
cities_final2.shape
```

```python
cities_final2.head()
```

```python
cities_final2.shape
```

```python
cities_final2['City'].nunique()
```

```python
cities_final3 = cities_final2.drop_duplicates(subset='City')
```

```python
cities_final3.shape
```

```python
#downloaded noaa weather database from google big query for weather data for 2019.
noaa = pd.read_csv('noaa_2019.csv')
```

```python
noaa.head()
```

```python
noaa.shape
```

```python
#data cleaning.
noaa['stn'].astype(int)
```

```python
noaa['stn'].nunique()
```

```python
noaa.info()
```

```python
cities_final3.head()
```

```python
cities_final3.info()
```

```python
cities_final3 = cities_final3.astype({'StationNumber':int})
```

```python
pd.set_option('display.max_columns', None)
```

```python
#combining weather data for relevant cities / stations for the year 2019. Only a single year data being used for model simplicity.
final_df = pd.merge(cities_final3, noaa,  how='left', left_on=['StationNumber'], right_on = ['stn'])
```

```python
final_df.head()
```

```python
final_df.shape
```

```python
#data cleaning and dropping unnecessary features.
```

```python
final_df = final_df.drop(['dewp','count_dewp','slp','count_slp','stp','count_stp','mxpsd','gust','max','flag_max','min','flag_min','flag_prcp','Lat','Lon','Address','count_temp','count_visib','count_wdsp','hail','thunder','tornado_funnel_cloud'], axis=1)
```

```python
final_df.shape
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
final_df['StationNumber'].nunique()
```

```python
final_df.isna().sum()
```

```python
final_df = final_df.dropna()
```

```python
final_df.shape
```

```python
final_df.isna().sum()
```

```python
final_df.head()
```

```python
final_df = final_df.drop('sndp', axis=1)
```

```python
final_df.head()
```

```python
import datetime as dt
```

```python
#converting day/ mo / year data into timestamps.
final_df['year'] = pd.to_datetime(final_df['year'], format='%Y').dt.year
```

```python
final_df['mo'] = pd.to_datetime(final_df['mo'], format='%m').dt.month
```

```python
final_df['da'] = pd.to_datetime(final_df['da'], format='%d').dt.day
```

```python
final_df.head()
```

```python
final_df['full_date'] = pd.to_datetime(final_df.year.astype(str)+'-'+final_df.mo.astype(str)+'-'+final_df.da.astype(str))
```

```python
final_df.head()
```

```python
final_df.set_index('full_date')
```

```python
final_df['City,State']  = final_df.City.astype(str)+ ',' + final_df.State.astype(str)
```

```python
final_df.head()
```
